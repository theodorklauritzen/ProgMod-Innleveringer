# Numerisk derivasjon

```python
import pylab

def f(x):
    return x ** 3 + 2 * x ** 2 - 5*x

def derivert(f, x, dx = 1e-8):
    return (f(x + dx) - f(x)) / dx

px = pylab.linspace(-3, 3, 100)
pfy = f(px)
pdfy = derivert(f, px)

pylab.axhline(y=0,color="black")
#pylab.axvline(x=0, color="red")
pylab.grid()

pylab.plot(px, pfy)
pylab.plot(px, pdfy)

pylab.show()
```
